### About Dataset:
#### Context
This dataset captures key health, economic, social, and immunization factors across 193 countries from 2000 to 2015, providing a robust foundation for analyzing factors impacting life expectancy. While past studies have included demographic and income variables, this dataset broadens the scope by incorporating immunization rates and the Human Development Index (HDI). With a focus on vaccinations (e.g., Hepatitis B, Polio, and Diphtheria), mortality rates, economic indicators, and other health-related factors, the data enables each country to identify critical factors contributing to lower life expectancy and implement targeted improvements for enhancing population longevity.

#### Content
The dataset, compiled from the World Health Organization (WHO) and United Nations (UN) data, consists of health and economic metrics critical to assessing life expectancy. Data from 193 countries has been organized by year from 2000 to 2015, yielding a total of 22 columns and 2938 rows. Predictive features are categorized into immunization, mortality, economic, and social factors. Some columns contain missing values, mainly for population, Hepatitis B, and GDP in less documented countries (e.g., Vanuatu, Tonga, Togo). These countries were excluded to maintain data quality and accuracy. Key columns include:

#### Columns:
- **Country**: Name of the country
- **Year**: Year of data collection
- **Status**: Developed or Developing status of the country
- **Life expectancy**: Average life expectancy in years
- **Adult Mortality**: Mortality rate for adults (per 1000 population)
- **Infant deaths**: Number of infant deaths per 1000 births
- **Alcohol**: Per capita alcohol consumption (in liters)
- **Percentage expenditure**: Government health expenditure as a percentage
- **Hepatitis B**: Hepatitis B immunization coverage (percentage) for one-year-olds
- **Measles**: Number of reported measles cases
- **BMI**: Average Body Mass Index
- **Under-five deaths**: Number of deaths in children under five (per 1000)
- **Polio**: Polio immunization coverage (percentage) for one-year-olds
- **Total expenditure**: Government health expenditure per capita
- **Diphtheria**: Diphtheria immunization coverage (percentage) for one-year-olds
- **HIV/AIDS**: Mortality rate due to HIV/AIDS
- **GDP**: Gross Domestic Product per capita
- **Population**: Population of the country
- **Thinness 1-19 years**: Percentage of thinness among 1-19-year-olds
- **Thinness 5-9 years**: Percentage of thinness among 5-9-year-olds
- **Income composition of resources**: HDI income composition index
- **Schooling**: Average number of years of schooling

This dataset serves as a rich source for building models to understand and predict life expectancy based on key health and socioeconomic indicators.

The BMI column contains several incorrect values, with BMIs recorded below 12 and above 60—even reaching values above 80—which are not realistic. However, since BMI is a crucial factor affecting life expectancy, we need to address these outliers through a more accurate imputation approach.

```
bmi_data = df.loc[(df['BMI'] < 12) | (df['BMI'] > 60), 'BMI'].groupby(df['Country'])

print(bmi_data.count().min())
print(bmi_data.count().max())
print(bmi_data.count().mean())
print(bmi_data.count().median())
```

```
#Replace invalid BMI values with NaN
df['BMI'] = np.where((df['BMI'] < 12) | (df['BMI'] > 60), np.nan, df['BMI'])
```

```
# Calculate median by country
bmi_medians = df.loc[df['BMI'].notna()].groupby('Country')['BMI'].median()
print(df['BMI'].isna().sum())

df['BMI'] = df['BMI'].fillna(df['Country'].map(bmi_medians))
print(df['BMI'].isna().sum())

# For countries that have no BMI data at all, fill remaining NaN values with a general median
general_median = df.loc[df['BMI'].notna()]['BMI'].median()
df['BMI'] = df['BMI'].fillna(general_median)
print(df['BMI'].isna().sum())
```

## spliting dataset into dependant and independant variables

```
x = df.drop('Life_expectancy', axis=1)
y = df['Life_expectancy']
```

```
x.head()
```

```
y.head()
```

## Encoding categorical data

```
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
x_encoded =x.copy()

cols_to_encode = ['Country', 'Status']
for col in cols_to_encode:
    x_encoded[col] = label_encoder.fit_transform(x[col])

print(x_encoded['Country'].unique())
print(x_encoded['Status'].unique())
```

## Feature Scaling

```
from sklearn.preprocessing import StandardScaler


x_scaled = x_encoded.copy()
scaler_x = StandardScaler()
x_scaled[x_encoded.columns] = scaler_x.fit_transform(x_encoded)
x_scaled.head()
```

## Splitting the dataset into the Training set and Test set

```
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size = 0.2, random_state = 0 )
```

## Multiple Linear Regression model

```
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

### Predicting the Test set results

```
y_pred = pd.DataFrame(regressor.predict(x_test).round(2), columns=['Predicted'])
y_pred['Actual'] = y_test.values

y_pred.head()
```

### Accuracy Score

```
regressor.score(x_test, y_test)
```

```
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred.iloc[:, 0])
mse
```

```
np.sqrt(mse)
```

###Conclusion:

The model achieved an accuracy score of 0.83, indicating it can explain around 83% of the variability in life expectancy based on the given features. Additionally, the root mean squared error (RMSE) is approximately 3 years. This means that the model's predictions for life expectancy are, on average, about 3 years off from the actual values. Given the range and natural variation in life expectancy across different countries, an RMSE of 3 years is reasonably good and suggests that the model is capturing general trends effectively. However, further improvements in feature selection or model refinement could help improve the precision for more accurate predictions.
